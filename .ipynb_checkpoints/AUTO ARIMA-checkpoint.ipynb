```python
def adfullertest(appl):
    result=adfuller(appl)
    labels= ['ADF Test statistics', 'p-value', '#lags used', 'number of observations used']
    for value, label in zip(result, label):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print ('strong evidence against the null hypothesis(Ho), reject the null hypothesis')
    else:
        print('weak evidence against null hypothesis, time series has a unit root, indoctaing it is non-staionary')
```

## SARIMA Model

```python
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook
```

```python
import itertools
def optimize_SARIMA(parameters_list, d,D,s,exog):
    """
    Return dataframe with parameters, corresponding AIC and SSE
    parameters_list - list with (p, q, P, Q) tuples
        d - integration order
        D - seasonal integration order
        s - length of season
        exog - the exogenous variable
    """
    results= []
    for param in tqdm_notebook(parameters_list):
        try:
            model_sarima= SARIMAX(train, order=param)
            model_sarima_fit= model_sarima.fit()
            #print(param,model_sarima_fit.aic)
        except:
            continue
        
        aic= model_sarima.aic
        results.append([param, aic])
        result_df=pd.DataFrame(results)
        result_df.coulmns= ['(p,q)*(P,Q)', 'AIC']
        
        #sort in ascending order
        result_df= result_df.sort_values(by='AIC', ascending= True).reset_index(drop=True)

        return result_df
        
        
#the a information criterion (AIC) is an estimator of in-sample prediction error and thereby relative quanity of 
# statistical models for a given set of data
# it's like the mean squared error in regression - the smaller the number the beteer the model
        
```

```python
result_df= optimize_SARIMA(parameters_list, 4,0,0, train['appl'])
```
