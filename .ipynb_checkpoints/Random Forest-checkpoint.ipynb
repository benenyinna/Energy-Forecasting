```python
edata = pd.read_csv("energydata.csv")
```

```python
edata['date']= pd.to_datetime(edata['date'], infer_datetime_format=True)
```

```python
edata= edata.set_index('date').resample('H').mean().reset_index() #aggregating into hourly basis
```

```python
edata['hour']=edata['date'].dt.hour
edata['dayofweek']=edata['date'].dt.dayofweek
edata['month']=edata['date'].dt.month
edata['year']=edata['date'].dt.year
edata['weekofyear']=edata['date'].dt.weekofyear
```

```python
#edata.set_index('date', inplace=True)
```

```python
edata.drop(['rv1','rv2'], axis=1, inplace=True)
```

```python
edata2=edata[['date', 'appl', 'lgt', 'temp1', 'rh_1', 'temp2', 'rh_2', 'temp3',
       'rh_3', 'temp4', 'rh_4', 'temp5', 'rh_5', 'temp6', 'rh_6', 'temp7',
       'rh_7', 'temp8', 'rh_8', 'temp9', 'rh_9', 'temp_out', 'press', 'rh_out',
       'windspeed', 'visibility', 'dewpoint', 'hour', 'dayofweek', 'month',
       'year', 'weekofyear']]
```

```python
edata2.drop(['date'], axis=1, inplace=True)
```

```python
y= edata2['appl'].values
```

```python
x= edata2.drop(labels=['appl'], axis=1)
```

```python
#extracting feature names for use later
feature_names= np.array(x.columns)
```

```python
scaler =StandardScaler()
scaler.fit(x)
x=scaler.transform(x)
```

```python
#splititing into test and train data
X_train, X_test, y_train, y_test= train_test_split(x,y, test_size=0.25, random_state=False)
```

```python
model=RandomForestRegressor()
```

```python
#using boruta
feat_selector= BorutaPy(model, n_estimators='auto', verbose=2, random_state=1)
```

```python
feat_selector.fit(X_train, y_train)
```

```python
print(feat_selector.support_) # shows whether a feature is meant to be selected or not
```

```python
print(feat_selector.ranking_) # ranking shows whether it should be accepted with 1 being the best
```

```python
X_filtered=feat_selector.transform(X_train)
```

```python
feature_ranks=list(zip(feature_names, feat_selector.ranking_, feat_selector.support_))
```

```python
for feat in feature_ranks:
    print('Feature:{:<31} Rank:{}, Keep:{}'.format(feat[0], feat[1], feat[2]))
```

```python
rand_model=RandomForestRegressor()
```

```python
rand_model.fit(X_filtered, y_train)
```

```python
X_test_filtered= feat_selector.transform(X_test)
prediction_ran= rand_model.predict(X_test_filtered)
```

```python
score= np.sqrt(mean_squared_error((y_test, prediction_ran))
print(f'RMSE Score on test set:{score:0.2f}')
```

```python
from sklearn.pipeline import make_pipeline

model= make_pipeline(StandardScaler(), RandomForestRegressor(n_estimators=1000, random_state=1))
model.fit(X_train, y_train)
print(model.score(X_train, y_train))
```

```python
feature_names = model[:-1].get_feature_names_out()

mdi_importances = pd.Series(
    model[-1].feature_importances_, index=feature_names
).sort_values(ascending=True)
```

```python
ax = mdi_importances.plot.barh()
ax.set_title("Random Forest Feature Importances (MDI)")
ax.figure.tight_layout()
```

```python
y_pred= model.predict(x_val)
print(model.score(x_val, y_val))
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

rms= sqrt(mean_squared_error(y_val, y_pred))
print(rms)
```

```python
df_val= pd.DataFrame({'Actual':y_val, 'Predicted':y_pred})
df_val
```

```python
##plotting predcitions
plt.plot(y_train)
plt.plot([None for i in y_train] + [x for x in y_val])
plt.plot([None for i in y_train] + [x for x in y_pred])
plt.show()
```
