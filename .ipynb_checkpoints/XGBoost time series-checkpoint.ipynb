```python
def create_features(ed_t):
    """
    create time series features based on time series index
    """
    ed_t['hour']=ed_t.index.hour
    ed_t['dayofweek']= ed_t.index.dayofweek
    ed_t['month']=ed_t.index.month
    ed_t['year']=ed_t.index.year
    ed_t['dayofmonth']=ed_t.index.day
    ed_t['weekofyear']=ed_t.index.isocalendar().week.astype(np.int64)
    return ed_t

ed_t=create_features(ed_t)
```

```python
ed_t.tail()
```

## Adding lag features

```python
def add_lags(ed_t):
    target_map= ed_t['appl'].to_dict()
    ed_t['lag1']=(ed_t.index- pd.Timedelta('30 days')).map(target_map)
    ed_t['lag2']=(ed_t.index- pd.Timedelta('60 days')).map(target_map)
    ed_t['lag3']=(ed_t.index- pd.Timedelta('90 days')).map(target_map)
    ed_t['lag4']=(ed_t.index- pd.Timedelta('120 days')).map(target_map)
    #edata['lag5']=(edata.index- pd.Timedelta('150 days')).map(target_map)
    return ed_t
```

```python
ed_t=add_lags(ed_t)
```

```python
tss= TimeSeriesSplit(n_splits=5)
ed_t= ed_t.sort_index()

fold=0
preds= []
scores=[]
for train_idx, val_idx in tss.split(edata):
    train1= ed_t.iloc[train_idx]
    test1= ed_t.iloc[val_idx]
    
    train1= create_features(train1)
    test1= create_features(test1)
    
    features_1= [ 'lgt', 'temp1', 'rh_1', 'temp2', 'rh_2', 'temp3', 'rh_3',
       'temp4', 'rh_4', 'temp5', 'rh_5', 'temp6', 'rh_6', 'temp7', 'rh_7',
       'temp8', 'rh_8', 'temp9', 'rh_9', 'temp_out', 'press', 'rh_out',
       'windspeed', 'visibility', 'dewpoint', 'hour', 'dayofweek', 'month',
       'year', 'dayofmonth', 'weekofyear', 'lag1', 'lag2', 'lag3', 'lag4']
    target_1= ['appl']
    
    X_train=train1[features_1]
    y_train=train1[target_1]
    
    X_test= test1[features_1]
    y_test= test1[target_1]
    
    reg= xgb.XGBRegressor(base_score=0.5, booster='gbtree', n_estimator=1000, early_stopping_rounds=50, objective='reg:linear', max_depth=3, learning_rate=0.01)
    reg.fit(X_train, y_train, eval_set=[(X_train, y_train), (X_test, y_test)], verbose=True)
    
y_pred=reg.predict(X_test)
preds.append(y_pred)
score_1= np.sqrt(mean_squared_error(y_test, y_pred))
scores.append(score_1)
```

```python
print(f'score across folds{np.mean(scores):0.4f}')
print(f'folds{np.mean(scores):0.4f}')
```
