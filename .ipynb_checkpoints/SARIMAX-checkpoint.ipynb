```python
mod = sm.tsa.statespace.SARIMAX(edata_s,
                                order=(3, 0, 1),
                                seasonal_order=(0, 0, 1, 7),
                                enforce_stationarity=False,
                                enforce_invertibility=False)
results = mod.fit()

results.plot_diagnostics(figsize=(16, 8))
plt.show()
```

## perfroming validation

```python
start_forecast = 2303
pred = results.get_prediction(start=start_forecast, dynamic=False)
pred_ci = pred.conf_int()

ax = edata_s.plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='Predictions')

ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)

ax.set_xlabel('Date')
ax.set_ylabel('Appliance Energy consumption')
plt.legend()

plt.show()
```

```python
## Creating time series for 1 and 5 months
```

```python
#%% Forecast 1m
days = 30
pred_uc = results.get_forecast(steps=30*days, dynamic=False)
pred_ci = pred_uc.conf_int()
ax = edata_s['appl'].plot(label='observed', figsize=(14, 7))
pred_uc.predicted_mean.plot(ax=ax, label='Forecast')
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.25)
ax.set_xlabel('Date')
ax.set_ylabel('Appliance Energy')
plt.legend()
plt.show()
#plt.savefig( + os.sep + 'SARIMAX_FORECAST_1y.png')
#plt.close()
```

```python
#%% Forecast 3m
#months = 3
pred_uc = results.get_forecast('2016-08-31', dynamic=False)
pred_ci = pred_uc.conf_int()
ax = edata_s.plot(label='observed', figsize=(14, 7))
pred_uc.predicted_mean.plot(ax=ax, label='Forecast')
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.25)
ax.set_xlabel('Date')
ax.set_ylabel('Energy')
plt.legend()
plt.show()
#plt.savefig(current_dir + os.sep + 'SARIMAX_FORECAST_3m.png')
plt.close()
```
