## Running a dickerfuller test

```python
# run ADF test
from statsmodels.tsa.stattools import adfuller

def adf_test(series,title=''):
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC') 
    
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)

    for key,val in result[4].items():
        out[f'critical value ({key})']=val
        
    print(out.to_string())          
    
    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("Data has no unit root and is stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis")
        print("Data has a unit root and is non-stationary")
        
# use ADF function       
adf_test(ed_h['appl'])
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(train_l)
scaled_train = scaler.transform(train_l)
scaled_test = scaler.transform(test_l)
# scaled_train = train
# scaled_test = test
```

```python
# Let's define to get 1176 Days back wich represents exactly one week and then predict the next week out
n_input = 1176
n_features = 1
generator = TimeseriesGenerator(scaled_train, scaled_train, length=n_input, batch_size=10)
```

```python
# Define Simple LSTM model
# define model
model = Sequential()
model.add(LSTM(128, activation='relu', input_shape=(n_input, n_features)))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
model.fit_generator(generator,epochs=30, shuffle=False)
```

```python
# model performance
model.history.history.keys()
loss_per_epoch = model.history.history['loss']
plt.plot(range(len(loss_per_epoch)),loss_per_epoch)
```
